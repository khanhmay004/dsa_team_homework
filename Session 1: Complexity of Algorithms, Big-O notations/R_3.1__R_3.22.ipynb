##### R-3.1 Graph the functions 8n, 4nlogn, 2n2, n3, and 2n using a logarithmic scale for the x- and y-axes; that is, if the function value f(n) is y, plot this as a point with x-coordinate at logn and y-coordinate at logy

```python
def plot_loglog():
    xs = range(1, 100)
    plt.loglog(xs, [8*x for x in xs], base = 2)
    plt.loglog(xs, [4*x*np.log2(x) for x in xs] , base = 2)
    plt.loglog(xs, [2*(x**2) for x in xs] , base = 2)
    plt.loglog(xs, [x**3 for x in xs] , base = 2)
    plt.loglog(xs, [2**x for x in xs] , base = 2)
    plt.legend(['$8n$', '$4n \cdot logn$', '$2n^2$', '$n^3$', '2^n'])
    plt.show()

plot_loglog()
```
