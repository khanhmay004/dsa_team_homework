#### R-3.23 - R-3.27 Give a big-Oh characterization, in terms of n, of the running time of the example1, example2,example3, example4, example5 function shown in Code Fragment 3.10

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def example1(S):
    """Return the sum of the elements in sequence S."""
    n = len(S)
    total = 0
    cnt = 0
    for j in range(n):
        total += S[j]
        cnt += 1
    return total, cnt

def example2(S):
    """Return the sum of the elemnets in sequnce S."""
    n = len(S)
    total = 0
    cnt = 0
    for j in range(0, n, 2):
        total += S[j]
        cnt += 1
    return total, cnt

def example3(S):
    """Return the sum of the elemnets in sequnce S."""
    n = len(S)
    total = 0
    cnt = 0
    for j in range(n):
        for k in range(1+j):
            total += S[j]
            cnt += 1
    return total, cnt

def example4(S):
    """Return the sum of the elemnets in sequnce S."""
    n = len(S)
    prefix = 0
    total = 0
    cnt = 0
    for j in range(n):
        prefix += S[j]
        total += prefix
        cnt += 1
    return total, cnt

def example5(A, B):
    n = len(A)
    count = 0
    num_cnt = 0
    for i in range(n):
        total = 0
        for j in range(n):
            for k in range(1+j):
                total += A[k]
                num_cnt += 1
        if B[i] == total:
            count+=1
    return count, num_cnt
```

First, we analyze the complexity of the previous code. In order to visualize it, we define the following function

```python
def time_complexity1(n, func):
    set_of_points = []
    for i in range(1, n):
        A = np.random.randint(0, 100, size=(i, ))
        _, num_cnt = func(A)
        set_of_points.append(num_cnt)
    fig = plt.figure(figsize=(8, 6), dpi= 80)
    plt.plot(set_of_points)
    plt.title(func.__name__)
    plt.show()
```

The time complexity of example1 is $O(n)$

```python
time_complexity1(100, example1)
```

The time complexity of example5 is $O(n^3)$
For example5 we define another visualization function:

```python
def time_complexity2(n = 100):
    set_of_points = []
    for i in range(1, n):
        A = np.random.randint(0, 100, size=(i, ))
        B = np.random.randint(0, 100, size=(i, ))
        _, num_cnt = example5(A, B)
        set_of_points.append(num_cnt)
    fig = plt.figure(figsize=(8, 6), dpi= 80)
    plt.plot(set_of_points)
    plt.plot(range(n), [i**2 for i in range(n)])
    plt.plot(range(n), [i**3 for i in range(n)])
    plt.legend(['example5', '$n^2$', '$n^3$'])
    plt.title('example5')
    plt.show()
```
